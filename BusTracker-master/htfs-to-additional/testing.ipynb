```python
import pandas as pd 
import numpy as np
```

```python
routes = pd.read_csv("updateRoutes.csv") 
routes
```

```python
calender = pd.read_csv("calender.csv") 
calender
```

```python
routeNcal = pd.merge(routes, calender, on='service_id', how="left")
routeNcal
```

```python
calender_date = pd.read_csv("calender_dates.csv") 
calender_date
```

```python
routeNcal = pd.merge(routeNcal, calender_date, on='service_id', how="left")
routeNcal
```

```python
routeNcal.to_csv('timetables.txt', header=True, index=False, sep=',', mode='a')
```
